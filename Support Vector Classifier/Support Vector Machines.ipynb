```python
#Iris dataset to classify various species of Iris flower
#Importing the libraries
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#To visualize how an Iris Setosa looks like
from IPython.display import Image
url = 'http://upload.wikimedia.org/wikipedia/commons/5/56/Kosaciec_szczecinkowaty_Iris_setosa.jpg'
Image(url,width=500,height = 500)
```

```python
#Loading the Iris dataset
iris = sns.load_dataset('iris')
```

```python
#View the first few rows of the dataset
iris.head()
```

```python
#Visualize correlation between various variables
sns.pairplot(iris,hue='species')
```

```python
#Kernel density estimate plot to see relation between sepal width and length for setosa
setosa = iris[iris['species']=='setosa']
sns.kdeplot(setosa['sepal_width'],setosa['sepal_length'])
```

```python
#Kernel density estimate plot to see relation between petal width and length for setosa
setosa = iris[iris['species']=='setosa']
sns.kdeplot(setosa['petal_width'],setosa['petal_length'])
```

```python
from sklearn.model_selection import train_test_split
```

```python
#Dividing predictor and predicted variables
X = iris.drop('species',axis=1)
y = iris['species']
```

```python
#split the dataset
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size = 0.3, random_state = 101)

#Importing SVC Model
from sklearn.svm import SVC
```

```python
model = SVC()
```

```python
#Fitting the SVC Model on training dataset
model.fit(X_train,y_train)
```

```python
#Predict on test dataset
predict = model.predict(X_test)
```

```python
#Importing classification report and confusion matrix function to assess accuracy
from sklearn.metrics import classification_report,confusion_matrix
```

```python
print(classification_report(y_test,predict))
```

```python
print(confusion_matrix(y_test,predict))
```
